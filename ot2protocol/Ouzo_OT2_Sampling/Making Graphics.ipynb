```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# For by OT2 uv vis
ethanol = np.arange(0,1,0.05)
absorbance = np.asarray([0.01,0.05,0.1,0.14,0.22,0.34,0.48,0.52,0.54,0.51,0.45,0.40,0.39,0.31,0.27,0.15,0.12,0.11,0.05,0])*0.2
fig, ax = plt.subplots(1)
ax.scatter(ethanol,absorbance)
ax.set_xlabel('Ethanol wtf')
ax.set_ylabel('AU')
plt.axis('scaled')
```

```python
# For Hand UV-Vis
ethanol = np.arange(0,1,0.05)
absorbance = np.asarray([0.01,0.05,0.1,0.14,0.22,0.34,0.48,0.52,0.54,0.51,0.45,0.40,0.39,0.31,0.27,0.15,0.12,0.11,0.05,0])*0.2
fig, ax = plt.subplots(1)
noise = np.random.normal(0,0.005,len(absorbance))
noisy_absorbance = absorbance+noise
ax.scatter(ethanol,noisy_absorbance)
ax.set_xlabel('Ethanol wtf')
ax.set_ylabel('AU')
```

```python
# For OT2 DLS, expect an increase in DLS size quality then a drop
ethanol_DLS = np.arange(0,1,0.05)
absorbance_DLS = np.asarray([0.01,0.15,0.18,0.22,0.23,0.34,0.48,0.52,0.54,0.54,.58,.60,1.39,2.31,2.27, 2.15,2.12,2.11,0.05,0])*650
fig, ax = plt.subplots(1)
noise = np.random.normal(0,100,len(absorbance_DLS))
noisy_absorbance = absorbance_DLS+noise
ax.scatter(ethanol_DLS,noisy_absorbance)
ax.set_xlabel('Ethanol wtf')
ax.set_ylabel('Size (nm)')
```

```python
# For Hand DLS, expect an increase in DLS size quality then a drop
ethanol_DLS = np.arange(0,1,0.05)
absorbance_DLS = np.asarray([0.01,0.15,0.18,0.22,0.23,0.34,0.48,0.52,0.54,0.54,.58,.60,1.39,2.31,2.27, 2.15,2.12,2.11,0.05,0])*650
fig, ax = plt.subplots(1)
ax.scatter(ethanol_DLS,absorbance_DLS)
ax.set_xlabel('Ethanol wtf')
ax.set_ylabel('Size (nm)')
```
