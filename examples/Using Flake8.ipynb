# Flake8 

### What is Flake8
Flake8 is wrapper function that checks a code against style guides (PEP8). It can ensure that every import is used, that comments are short or long enough, and that style is consistant throughout. This is very useful for collaberations and situations where multiple people need to ensure that the genral style is consitant throughout the entire repository.

### Flake8 Installation
Flake8 needs to be installed on the computer before it can be run. 


1) Using Anaconda Prompt: 

```conda install -c anaconda flake8```

2) Using Pip Install: 

```pip install flake8```

## Command Line use of Flake8
Flake8 can be run directly from the command line with ```flake8``` run in the folder that you are looking to check. Or you can investigate a specific folder using ```flake8 file_name```.
